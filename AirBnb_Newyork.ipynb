```python
# Installing Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
data = pd.read_csv('Newyork_listing_2025.csv', encoding_errors='ignore')
```

```python
data.head()
```

```python
data.tail()
```

```python
data.shape
```

```python
data.info()
```

```python
#Statistical Summary
data.describe()
```

```python
# Data Cleaning 
data.isnull().sum()

# Dropping all missing values rows
data.dropna(inplace=True)

# If incase replace null values
# data.fillna()

data.isnull().sum()
```

```python
data.shape
```

```python
# Dealing with duplicate rows
data.duplicated().sum()

# Filtering the duplicated values
data[data.duplicated()]

# deleting all the duplicated rows
data.drop_duplicates(inplace=True)

data.duplicated().sum()
```

```python
# Checking data type
data.dtypes

# Changing the data type of the variables 
data['id']= data['id'].astype(object)

data['host_id']= data['host_id'].astype(object)
data.dtypes
```

```python
# Data Analysis
#Univariate Analysis
# Identifiying outliers in price
sns.boxplot(data=data, x='price')
```

```python
df = data[data['price'] <1500]
sns.boxplot(data= df, x='price')
```

```python
# Price Distribution 

plt.figure(figsize=(8, 5))
sns.histplot(data= df, x='price', bins=100)
plt.title('Price Distribution')
plt.ylabel("Frequency")
plt.show()
```

```python
df.columns
df.dtype
```

```python
# availability_365 Distribution 

plt.figure(figsize=(10, 4))
sns.histplot(data= df, x='availability_365', bins=100)
plt.title('availability_365 Distribution')
plt.ylabel("Frequency")
plt.show()
```

```python
data.dtypes
```

```python
#Using group by

df.groupby(by='neighbourhood_group')['price'].mean()
```

```python
df.head()
```

```python
#Adding a new column (price per bed)

df['price per bed']=df['price']/df['beds']
df.head()
```

```python
#Average price per bed

df.groupby(by='neighbourhood_group')['price per bed'].mean()
```

```python
#Bi Variable Analysis (One variable dependency with another variable)

df.columns
```

```python
# Price dependency on neighbourhood
sns.barplot(data= df, x='neighbourhood_group', y='price', hue='room_type')
```

```python
# NUmber of reviews and price relationship

plt.title("Locality and Review Dependency")
sns.scatterplot(data= df, x='number_of_reviews', y='price', hue='neighbourhood_group')
```

```python
df.dtypes
```

```python
sns.pairplot(data= df, vars=['price','minimum_nights','number_of_reviews','availability_365'], hue='room_type')
```

```python
#Geographical Distribution of AirBnb Listing
plt.figure(figsize=(10,8))
plt.title('Geographical Distribution of AirBnb Listing')
sns.scatterplot(data=df, x='longitude', y='latitude', hue='room_type')
plt.show()
```

```python
df.dtypes
```

```python
# Heatmap - Correlation of one variable with others for numerical column
corr = df[['latitude','longitude','price','minimum_nights','number_of_reviews','reviews_per_month','availability_365','beds']].corr()
corr
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(data=corr, annot= True)
```
